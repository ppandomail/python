## Excepciones

### Creaci√≥n de la excepci√≥n

```python
class SinNaftaError(Exception):
    pass
```

### Lanzamiento de la excepci√≥n

```python
import random
nafta = None
def arrancar():
    nafta = random.randrange(1, 10)
    if nafta < 5:
        raise SinNaftaError()  
    print(nafta)
    
```

### Manejo de la excepci√≥n

```python
try:
    arrancar()
except SinNaftaError:
    print('Cargar el tanque')
```

## Archivos

```python
from io import open
```

### Escritura

```python
  # Si el archivo no existe, el m√©todo open lo crea
  # 'w': escritura -> reemplaza
  # 'a': append    -> agrega
  fw = open('/Users/ppando/Materias/python/doc/data/datos.txt', 'a')    
  fw.write('Hola mundo \ncruel')
  fw.close()
```

### Lectura

```python
  # 'r': lectura
  fr = open('/Users/ppando/Materias/python/doc/data/datos.txt', 'r')
  lista = fr.readlines()      # devuelve lista de lineas
  fr.seek(0)                  # se cambia posici√≥n del puntero al caracter en posici√≥n 0
  texto = fr.read()           # devuelve todo el contenido del archivo
  print(texto)
  print(lista)
  fr.seek(0)                  
  print(fr.read(11))          # lee hasta el caracter en posici√≥n 11
  fr.close()
```
